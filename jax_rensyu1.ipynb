```python
import jax.numpy as np
import jax
```

```python
def f(x):
  return np.cos(x)

#関数の定義

print(f(np.pi))
```

```python
print(f(np.array([np.pi/2, np.pi])))
#配列を渡す
```

```python
grad_f = jax.grad(f)
#関数を微分

print(grad_f(np.pi))
```

```python
grad_grad_f = jax.grad(grad_f)
#もう一度微分

print(grad_grad_f(np.pi))
```

```python
print(jax.vmap(grad_f)(np.array([np.pi, np.pi/2])))

#jax.vmapを使えば配列を渡せる
```

```python
jit_vmap_grad_f = jax.jit(jax.vmap(grad_f))
#jitで高速化

print(jit_vmap_grad_f(np.array([np.pi, np.pi/2])))
```

```python
def g(x):
  x,y =x
  return x**2 + 2*x*y + 2*y**2

print(jax.hessian(g)(np.array([1.0, 1.0])))
#ヘッセ行列の計算
```

```python
print(jax.jacobian(g)(np.array([1.0, 2.0])))
#ヤコビ行列の計算
#[2x+2y, 2x+4y]
```
